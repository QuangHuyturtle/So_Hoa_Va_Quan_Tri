```python
# bài tập 2
import numpy as np

random_array = np.random.rand(20)
print("Mảng ngẫu nhiên từ 0 đến 1:", random_array)


min_val = random_array.min()
max_val = random_array.max()
normalized_array = (random_array - min_val) / (max_val - min_val)
print("Mảng sau khi chuẩn hóa:", normalized_array)

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Tích vô hướng của a và b:", dot_product)

matrix = np.random.randint(1, 10, size=(5, 5))
print("Ma trận:")
print(matrix)


determinant = np.linalg.det(matrix)
print("Định thức của ma trận:", determinant)

if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("Ma trận nghịch đảo:")
    print(inverse_matrix)
else:
    print("Ma trận không khả nghịch (định thức = 0).")
```

```python
#Bài tâp 3
import pandas as pd


data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [23, 25, 22, 24, 21],
    "Score": [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)

print("DataFrame ban đầu:")
print(df)


average_score = df["Score"].mean()
print("\nGiá trị trung bình của cột 'Score':", average_score)


filtered_df = df[df["Score"] > 85]
print("\nCác hàng có 'Score' lớn hơn 85:")
print(filtered_df)
```

```python
#Bài 4
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

x = np.linspace(-10, 10, 100)
y = x ** 2
plt.figure(figsize=(6, 4))
plt.plot(x, y, label="y = x^2", color="blue")
plt.title("Biểu đồ hàm số y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Score": [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)

plt.figure(figsize=(6, 4))
plt.bar(df["Name"], df["Score"], color="skyblue")
plt.title("Điểm số của sinh viên")
plt.xlabel("Tên")
plt.ylabel("Điểm số")
plt.show()

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_target_counts = iris_df['target'].value_counts()

labels = iris.target_names
sizes = iris_target_counts
colors = ['gold', 'lightcoral', 'skyblue']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Phần trăm mỗi loại hoa trong tập Iris")
plt.show()
```

```python
# Bài tập 7
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris


iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target


iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

plt.figure(figsize=(8, 6))

for species, color in zip(['setosa', 'versicolor', 'virginica'], ['red', 'blue', 'green']):
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'],
                label=species, color=color, alpha=0.7)

plt.title("Biểu đồ phân tán giữa sepal_length và sepal_width", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Sepal Width (cm)", fontsize=12)
plt.legend(title="Loại hoa", fontsize=10)
plt.grid(True)

plt.show()
```
