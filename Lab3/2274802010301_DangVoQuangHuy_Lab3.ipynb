```python
import pandas as pd
import numpy as np

data = {
    "ID": [101, 102, 103, 104, 105],
    "Name": ["Quang Huy", "Gia Huy", None, "Minh Tam", "Nguyen Tin"],
    "Age": [20, 22, None, 30, 40],
    "Salary": [50000, 10000, 65000, 21000, None]
}

df = pd.DataFrame(data)

# Điền giá trị thiếu trong cột Age bằng giá trị trung bình
df["Age"].fillna(df["Age"].mean(),inplace=True)

# Điền giá trị thiếu trong Name bằng "Unknown"
df["Name"].fillna("Unknown",inplace=True)

# Điền giá trị thiếu trong Salary bằng phương pháp Interpolation
df["Salary"].interpolate(methode="linear",inplace=True)

# Loại bỏ các dòng chứa dữ liệu thiếu
df.dropna(inplace=True)

# Hiển thị DataFrame sau khi xử lý
print(df)

```

```python
#cho data

df1=pd.DataFrame({
    "Id":[1,2,3],
    "Score_A":[70,90,85]
})

df2=pd.DataFrame({
    "Id":[3,4,5],
    "Score_B":[62,91,75]
})

#Thực hiện Merge trên cột id ( Inner join , left join, outer join)
# Inner Join (Chỉ giữ các giá trị chung giữa hai bảng)
df_inner = pd.merge(df1, df2, on="Id", how="inner")
print("Inner Join:\n", df_inner)

# Left Join (Giữ tất cả giá trị của df1, điền NaN nếu không khớp)
df_left = pd.merge(df1, df2, on="Id", how="left")
print("\nLeft Join:\n", df_left)

# Outer Join (Giữ tất cả giá trị của cả df1 và df2, điền NaN nếu không có)
df_outer = pd.merge(df1, df2, on="Id", how="outer")
print("\nOuter Join:\n", df_outer)
#Nối DataFrame theo chiều dọc 
pd.concat([df1,df2],axis=0)
#Gộp df1 và df2 để điền giá tri thiếu
df1.set_index("Id").combine_first(df2.set_index("Id")).reset_index()


```

```python
import pandas as pd
import numpy as np

# Tạo dữ liệu
data = pd.DataFrame({
    "ID": np.arange(1, 1000001, dtype=np.int32),  # Chỉnh sửa số lượng phần tử
    "Value": np.random.randint(1, 100, 1000000).astype(np.int8)  # Chuyển luôn khi tạo
})

# Kiểm tra mức độ sử dụng bộ nhớ
print(data.info())

# Tìm 5 giá trị phổ biến nhất trong cột Value
top_values = data["Value"].value_counts().head(5)
print(top_values)

# Lọc dữ liệu với query
filtered_data = data.query("Value > 90")
print(filtered_data)
```

```python
#Biểu đồ nhiều đường

#Cho dữ liệu sau

import matplotlib.pyplot as plt
import numpy as np

t=np.linspace(0,10,100)
y1=np.sin(t)
y2=np.cos(t)
y3=np.sin(t)*np.cos(t)

#tạo biểu đồ các đường theo thời gian
# Vẽ biểu đồ
plt.figure(figsize=(10, 5))  # Kích thước biểu đồ

plt.plot(t, y1, label="sin(t)", color="r", linestyle="-", linewidth=2)
plt.plot(t, y2, label="cos(t)", color="b", linestyle="--", linewidth=2)
plt.plot(t, y3, label="sin(t) * cos(t)", color="g", linestyle=":", linewidth=2)

# Thêm tiêu đề và nhãn
plt.title("Biểu đồ nhiều đường theo thời gian", fontsize=14)
plt.xlabel("Thời gian (t)", fontsize=12)
plt.ylabel("Giá trị", fontsize=12)
plt.legend()  # Hiển thị chú thích
plt.grid(True)  # Hiển thị lưới
plt.show()
```

```python
#Biểu đồ thanh nhóm
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
labels = ["Q1", "Q2", "Q3", "Q4"]
A = [500, 700, 800, 600]
B = [450, 350, 650, 750]
C = [500, 250, 850, 600]

# Tạo vị trí cho từng nhóm thanh
x = np.arange(len(labels))  # Tạo vị trí trên trục x
width = 0.2  # Độ rộng của mỗi thanh

# Vẽ biểu đồ thanh nhóm
plt.figure(figsize=(8, 5))
plt.bar(x - width, A, width, label="A", color="r")  # Thanh cho A
plt.bar(x, B, width, label="B", color="b")  # Thanh cho B
plt.bar(x + width, C, width, label="C", color="g")  # Thanh cho C

# Cài đặt nhãn và tiêu đề
plt.xlabel("Quý")
plt.ylabel("Doanh thu")
plt.title("Doanh thu của A, B, C theo quý")
plt.xticks(x, labels)  # Gán nhãn trục x
plt.legend()  # Hiển thị chú thích
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Thêm lưới ngang

# Hiển thị biểu đồ
plt.show()
```

```python
#Biểu đồ tròn
import matplotlib.pyplot as plt

# Dữ liệu
Cty = ["A", "B", "C", "D"]
thiphan = [30, 25, 25, 20]
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]  # Sửa màu HEX hợp lệ (có dấu #)

# Vẽ biểu đồ tròn
plt.figure(figsize=(6, 6))
plt.pie(thiphan, labels=Cty, colors=colors, autopct="%1.1f%%", startangle=140, wedgeprops={"edgecolor": "black"})

# Tiêu đề
plt.title("Tỷ lệ thị phần của 4 công ty")

# Hiển thị biểu đồ
plt.show()
```

```python
#Biểu đồ phân tán
import matplotlib.pyplot as plt
import numpy as np

# Tạo dữ liệu ngẫu nhiên
x = np.random.rand(100)  # 100 giá trị ngẫu nhiên cho trục x
y = np.random.rand(100)  # 100 giá trị ngẫu nhiên cho trục y
sizes = np.random.rand(100) * 300  # Kích thước điểm (scale lên 300)
colors = np.random.rand(100)  # Màu sắc ngẫu nhiên

# Vẽ biểu đồ phân tán
plt.figure(figsize=(7, 5))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, cmap="viridis", edgecolors="black")

# Thêm tiêu đề và nhãn
plt.title("Biểu đồ phân tán của 2 biến ngẫu nhiên", fontsize=14)
plt.xlabel("Biến X", fontsize=12)
plt.ylabel("Biến Y", fontsize=12)
plt.colorbar(label="Màu sắc")  # Thêm thanh màu thể hiện giá trị màu

# Hiển thị biểu đồ
plt.show()
```

```python
#Biểu đồ nhiệt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tạo dữ liệu ngẫu nhiên (ma trận 10x10)
data = np.random.rand(10, 10)

# Vẽ biểu đồ nhiệt
plt.figure(figsize=(8, 6))
sns.heatmap(data, cmap="coolwarm", annot=True, fmt=".2f", linewidths=0.5, cbar=True)

# Thêm tiêu đề
plt.title("Biểu đồ nhiệt của ma trận dữ liệu", fontsize=14)

# Hiển thị biểu đồ
plt.show()
```
